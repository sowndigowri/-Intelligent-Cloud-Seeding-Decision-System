```python
import os
import numpy as np
import tensorflow as tf

TRAIN_DIR = "data/raw/cloudcast/CloudCastSmall/TrainCloud"

# List all .npy files
files = sorted([os.path.join(TRAIN_DIR, f) for f in os.listdir(TRAIN_DIR) if f.endswith('.npy')])
print(f"Found {len(files)} .npy files.")

# Example: single class labels (all zeros)
labels = np.zeros(len(files))  # Change if multiple classes

# ----------------------------
# 1 Generator function
# ----------------------------
def npy_generator(files, labels):
    for f, label in zip(files, labels):
        arr = np.load(f)

        # Handle 1D arrays
        if arr.ndim == 1:
            side = int(np.sqrt(arr.shape[0]))
            arr = arr[:side*side].reshape((side, side))

        # Handle 2D arrays
        if arr.ndim == 2:
            arr = np.expand_dims(arr, axis=-1)

        tensor = tf.convert_to_tensor(arr, dtype=tf.float32)
        tensor = tf.image.resize(tensor, [128, 128])

        # Convert grayscale to RGB
        if tensor.shape[-1] == 1:
            tensor = tf.image.grayscale_to_rgb(tensor)

        yield tensor, label

# ----------------------------
# 2 Create tf.data.Dataset
# ----------------------------
dataset = tf.data.Dataset.from_generator(
    lambda: npy_generator(files, labels),
    output_signature=(
        tf.TensorSpec(shape=(128,128,3), dtype=tf.float32),
        tf.TensorSpec(shape=(), dtype=tf.float32)
    )
)

# ----------------------------
# 3 Shuffle, batch, prefetch
# ----------------------------
train_size = int(0.8 * len(files))
val_size = len(files) - train_size

train_ds = dataset.take(train_size).shuffle(1000).batch(32).prefetch(tf.data.AUTOTUNE)
val_ds   = dataset.skip(train_size).batch(32).prefetch(tf.data.AUTOTUNE)

print("Train batches:", len(list(train_ds)))
print("Validation batches:", len(list(val_ds)))

```

```python
from tensorflow.keras import layers, models

model = models.Sequential([
    layers.Input(shape=(128,128,3)),

    layers.Conv2D(32, (3,3), activation='relu'),
    layers.MaxPooling2D((2,2)),

    layers.Conv2D(64, (3,3), activation='relu'),
    layers.MaxPooling2D((2,2)),

    layers.Conv2D(128, (3,3), activation='relu'),
    layers.MaxPooling2D((2,2)),

    layers.Flatten(),
    layers.Dense(128, activation='relu'),
    layers.Dense(1, activation='sigmoid')  # Change to number of classes for multi-class
])

model.compile(
    optimizer='adam',
    loss='binary_crossentropy',  # Use 'sparse_categorical_crossentropy' if multi-class
    metrics=['accuracy']
)

model.summary()
```

```python
history = model.fit(
    train_ds,
    validation_data=val_ds,
    epochs=1 # Start small, increase later
)
```

```python

import os
import numpy as np
import pandas as pd
import tensorflow as tf
from tensorflow.keras import layers, models
from sklearn.ensemble import RandomForestRegressor
import joblib
from PIL import Image
import matplotlib.pyplot as plt
import ipywidgets as widgets
from IPython.display import display
```

```python
# Load weather dataset (update path if needed)
df = pd.read_csv("/content/weatherAUS.csv").dropna()

features = ["Humidity3pm", "Pressure3pm", "Temp3pm", "WindSpeed3pm"]
target = "Rainfall"

X = df[features]
y = df[target]

rain_model = RandomForestRegressor(
    n_estimators=150,
    random_state=42
)
rain_model.fit(X, y)

joblib.dump(rain_model, "rainfall_model.pkl")
print(" Rainfall model saved")
```

```python
cnn_model = models.Sequential([
    layers.Input(shape=(128,128,1)),
    layers.Conv2D(32, (3,3), activation="relu"),
    layers.MaxPooling2D(),
    layers.Conv2D(64, (3,3), activation="relu"),
    layers.MaxPooling2D(),
    layers.Flatten(),
    layers.Dense(128, activation="relu"),
    layers.Dense(4, activation="softmax")  # 4 cloud types
])

cnn_model.compile(
    optimizer="adam",
    loss="sparse_categorical_crossentropy",
    metrics=["accuracy"]
)

cnn_model.summary()
```

```python
def load_npy_image(path):
    img = np.load(path)

    # Normalize
    img = img.astype("float32")
    img = (img - img.min()) / (img.max() - img.min() + 1e-6)

    # Resize
    img = tf.image.resize(img[..., None], (128,128))

    return img.numpy()
```

```python
def cloud_seeding_decision(image, weather):

    img = image.resize((128,128))
    img = img.convert("L")
    img = np.array(img)/255.0
    img = img[..., None]
    img = img[None, ...]

    cloud_probs = cnn_model.predict(img, verbose=0)
    cloud_class = np.argmax(cloud_probs)

    rain_mm = rain_model.predict([weather])[0]

    print("DEBUG  Cloud class:", cloud_class)
    print("DEBUG  Rainfall (mm):", rain_mm)

    if cloud_class == 3 and rain_mm < 5:
        return " SEED NOW"
    elif cloud_class == 2:
        return " WAIT"
    else:
        return " DO NOT SEED"
```

```python
import ipywidgets as widgets
from IPython.display import display
from PIL import Image
import matplotlib.pyplot as plt
import io

upload = widgets.FileUpload(
    accept="image/*",
    multiple=False
)

h = widgets.FloatSlider(value=70, min=0, max=100, step=1, description="Humidity")
p = widgets.FloatSlider(value=1010, min=900, max=1100, step=1, description="Pressure")
t = widgets.FloatSlider(value=30, min=-10, max=50, step=1, description="Temp")
w = widgets.FloatSlider(value=10, min=0, max=60, step=1, description="Wind")

btn = widgets.Button(description="Predict", button_style="success")
out = widgets.Output()

def run_prediction(b):
    with out:
        out.clear_output()

        if not upload.value:
            print(" Upload an image first")
            return

        file = list(upload.value.values())[0]
        image = Image.open(io.BytesIO(file["content"]))

        decision = cloud_seeding_decision(
            image,
            [h.value, p.value, t.value, w.value]
        )

        plt.imshow(image)
        plt.axis("off")
        plt.show()

        print(" Decision:", decision)

btn.on_click(run_prediction)

display(
    widgets.VBox([
        upload,
        h, p, t, w,
        btn,
        out
    ])
)
```
